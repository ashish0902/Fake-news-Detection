## removing stopwords from total and lemmatizing them

```python
import re
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
```

```python
from nltk.stem import WordNetLemmatizer
lt = WordNetLemmatizer()
```

```python
import nltk
nltk.download('punkt')
```

```python
for index,row in x.iterrows():
    sentence = row['total']
    sentence = re.sub(r'[^\w\s]'," ",sentence)
    x.loc[index,'total'] = sentence
```

```python
for index,row in x.iterrows():
    sentence = row['total']
    x.loc[index,'total'] = sentence.lower()
```

```python
x.head(100)
```

```python
a = 'been had done language cities mice'
h = a.split()
b = ""
h = [w for w in h if not w in stop_words]
for word in h:
   b = b + " " + lt.lemmatize(word)
print(b)
```

```python
corpus = []
for index,row in x.iterrows():
    filter_sentence = ""
    sentence = row['total']
    words = sentence.split()
    
    words = [w for w in words if not w in stop_words]
    
    for word in words:
        filter_sentence = filter_sentence + " " +lt.lemmatize(word)
        
    corpus.append(filter_sentence)
```
